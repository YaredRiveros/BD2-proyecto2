```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd
import os
import pickle
```

# Construccion del indice invertido

```python
df_processed = pd.read_csv('DataTwitter_processed.csv')
```

```python
df_processed = df_processed.dropna(subset=['processed_text'])
```

```python
# Definimos la cantidad de tweets que vamos a procesar a la vez
block_size = 10000  # Ajusta este valor según la memoria disponible

# Creamos el vectorizador TF-IDF
vectorizer = TfidfVectorizer()

# Leemos los tweets procesados en bloques y los indexamos
for i in range(0, df_processed.shape[0], block_size):
    print(f"Procesando bloque {i // block_size + 1}...")
    
    # Tomamos el bloque actual de tweets
    block = df_processed.iloc[i:i+block_size]['processed_text'].dropna().values.astype('U')

    # Aplicamos el vectorizador a los tweets del bloque
    tfidf_matrix = vectorizer.fit_transform(block)

    # Obtenemos las palabras (features) y los pesos TF-IDF
    features = vectorizer.get_feature_names_out()
    tfidf_weights = tfidf_matrix.toarray()

    # Creamos el índice invertido para el bloque
    inverted_index_block = {}
    for j, weights in enumerate(tfidf_weights):
        for k, weight in enumerate(weights):
            if weight > 0:
                term = features[k]
                if term in inverted_index_block:
                    inverted_index_block[term].append((i+j, weight))
                else:
                    inverted_index_block[term] = [(i+j, weight)]
    
    # Guardamos el índice invertido del bloque
    with open(f"inverted_index_block_{i // block_size + 1}.pkl", "wb") as f:
        pickle.dump(inverted_index_block, f)
        
print("Terminado.")
```
